```python
#importing data on foreign born population in Germany from https://de.statista.com/statistik/daten/studie/71210/umfrage/auslaender-pro-bundesland/
gerfb=pd.read_excel('https://docs.google.com/uc?id=1V6ZoC0HvmVnGKbnXj98IE3QvCKNKNFfg&export=download')
```

```python
gerfb.head(5)
```

```python
gerfb.dtypes
```

```python
#limiting dataset only to region, county code, and total number of foreign born people per county
gerfb1=gerfb[['Region', 'AGS_Kreis', 'Geschlecht_Insgesamt_Anzahl']]
```

```python
gerfb1.head(5)
```

```python
#renaming county code column to name I will later use to merge data with map
gerfb2 = gerfb1.rename(columns={'AGS_Kreis': 'county_code'})
```

```python
gerfb2.columns
```

```python
ger0.columns
```

```python
#figuring out which columns of the map data contains county codes
ger0['krs_code']
```

```python
#trying to remove square brackets and quotation marks from county codes to allow merge
#code from https://stackoverflow.com/questions/38147447/how-to-remove-square-bracket-from-pandas-dataframe
ger0['krs_code'] = ger0['krs_code'].str.strip('[]')
ger0['county_code'] = ger0['krs_code'].str.replace("'", "")
```

```python
ger0['county_code']
```

```python
#making the county_code column in both data sets floats
#command for coercing if not all values are numbers from: https://www.geeksforgeeks.org/how-to-handle-pandas-value-error-could-not-convert-string-to-float/
ger0['county_code2'] = ger0['county_code'].astype(float)
gerfb2['county_code2'] = pd.to_numeric(gerfb2['county_code'], errors='coerce')
```

```python
#merging, code from map.ipynb by AOK
gerfb3 = pd.merge(ger0, gerfb2, on='county_code2',how='outer',indicator=True)
gerfb3[['Region','county_code2','_merge']]

gerfb_final = pd.merge(ger0, gerfb2, on='county_code2',how='inner',indicator=False)
```

```python
#making the map, using code from map.ipynb by AOK
#suggestion to use thousands by AOK
#code to place legend outside map by AOK
gerfb_final['Geschlecht_Insgesamt_Anzahl-k'] = gerfb_final['Geschlecht_Insgesamt_Anzahl']/1000

fig, ax = plt.subplots(1, figsize=(6,10))
gerfb_final.plot(ax=ax,column='Geschlecht_Insgesamt_Anzahl-k',legend=True,cmap='YlOrRd',
          scheme='natural_breaks',k=8, edgecolor='grey',linewidth=0.5,
          legend_kwds= {'bbox_to_anchor': (1, 1),"fmt": "{:,.0f}",'loc':'upper left', 'markerscale':1.2,'title_fontsize':'small','fontsize':'7'})
leg1 = ax.get_legend()
leg1.set_title('''white=missing''')

ax.title.set_text("Foreign Born Population in Germany (thousands)")
ax.set_xticks([]) #kills xticks
ax.set_yticks([]) #kills xticks
ax.set_xlabel('''https://de.statista.com/statistik/daten/studie/71210/umfrage/auslaender-pro-bundesland/''')
```

Interpretation of maps

These two maps show very clearly that the areas that had the most votes for the AFD in the 2021 general elections were areas that generally did not have a very high number of foreign born people. A particularily interesting case is Berlin, which stands out as an 'island' in both maps. On the one hand, it is one if the places in Germany with the highest numbers of foreign born people. On the other hand, the number of people who voted for the AFD in the 2021 general election is relatively low, especially compared to the surrounding areas.

This pattern of high volume of votes for the AFD in areas with lower numbers of foreign born population suggests, that people's decision to vote for the AFD is not necessarily influenced by having a lot of experiences with people who migrated to Germany. One explanation could be, that people in those areas have less contact with people who migrated to Germany, making it easier for the AFD to spread propaganda that dehumanizes people who migrated. What remains unanswered, however, is the question of what other factors make people in these areas so susceptible to far right extremist views. A map that could explore this question could, for example, vizualize the per capita income or the percentage of people living below the poverty line in those areas.

Some ways to improve the maps produced in this problem set would be to calculate the percentage of foreign born population instead of the total numbers for each county. The election resuts map could be improved by including other parties with anti-migration policies, as well as more up to date numbers.
